```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.themoviedb.org/movie"
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
source_data = requests.get(url,headers = header).text
```

```python
soup_data = BeautifulSoup(source_data,'lxml')
```

```python
all_url = []
for i in range(1, 51):
    all_page = f"?all_page={i}"
    all_url.append(url + all_page)
all_url
```

```python
movie_data = []
for i in all_url:
    source_movie_data = requests.get(i, headers = header).text
    soup_movie_data = BeautifulSoup(source_movie_data,'lxml')
    poster = soup_movie_data.find_all('div',class_='card style_1')
    
    for item in poster:
        inside_poster_link = str(item.find('a'))[29:37] 
        new_url = url + inside_poster_link
        picture = requests.get(new_url,headers = header).text
        soup_movie = BeautifulSoup(picture,'lxml')
        
        all_given_heading = soup_movie.find('section', id='original_header')
        
# MOVIE NAME        
        try:
            movie_name = all_given_heading.find('h2').find('a').text
        except:
            movie_name = None
            
# MOVIE RELEASE DATE        
        try:
            release_date = all_given_heading.find('span', class_='release').text.replace('\n','').strip().split(" ")[0]
        except:
            release_date = None 
        
#  MOVIE RATING        
        try:
            rating = all_given_heading.find('div', class_='user_score_chart')["data-percent"]
        except:
            rating = None
            
# MOVIE DIRECTOR NAME
        try:
            production = all_given_heading.find('ol', class_='people no_image')
            profile = production.find_all('li', class_='profile')
            director = ""
            for i in range(0, len(profile)):
                character = str(profile[i].find_all('p')[1].text)
                if 'Director' in character:
                    director+= f"{profile[i].find('a').text},"
            director = director.rstrip(',')
        except:
            director = None
            
# MOVIE DURATION        
        try:
            duration = all_given_heading.find('span', class_='runtime').text.replace('\n','').strip()
        except:
            duration = None 
        
# MOVIE GENRE         
        try:
            genre = all_given_heading.find('span', class_='genres').text.replace('\n','').replace('\xa0','')
        except:
            genre = None
        
# DATA STORED IN DICTIONARY FORM        
        full_movie_dict = {
            'Movie Name' : movie_name,
            'Release Date' : release_date,
            'Rating' : rating,
            'Director' : director,
            'Duration' : duration,
            'Genre' : genre
        }
        movie_data.append(full_movie_dict)
movie_data
```

```python
import pandas as pd
df = pd.DataFrame(movie_data)
df.to_excel('Moviedb_project.xlsx')
```
